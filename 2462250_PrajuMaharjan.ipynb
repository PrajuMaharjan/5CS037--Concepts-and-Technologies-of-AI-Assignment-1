```python
# Importing necessary libraries

import pandas as pd # For data analysis and manipulation
import matplotlib.pyplot as plt # For creating graphs and charts
import seaborn as sns # FOr advanced data visualizations
```

```python
# Reading csv file from folder in drive

hdi_df=pd.read_csv('/content/drive/MyDrive/5CS037--Concepts-and-Technologies-of-AI/CourseWork/Human_Development_Index_Dataset.csv',encoding='latin1')
```

```python
# 3 Problem 1 : Basic Data Exploration & Trend Visualization
# 3.1 Problem- 1A- Single Year HDI Exploration (Latest Year: 2022)
# Task-1 : Extract Latest Year

# Identification of unique years in the dataset and filtration of observations from the year 2022 into a new dataframe

print("Years in the dataset : ")
print(hdi_df["year"].unique())
hdi_2022_df=hdi_df[hdi_df["year"]==2022]
```

```python
# Task-2 : Data Exploration

# Displaying first 10 rows of the 2022 dataset
print('First 10 rows of filtered dataset : ')
display(hdi_2022_df.head(10))

# Displaying number of rows and columns in the 2022 dataset
print(f"The total number of rows in the filtered dataset is {hdi_2022_df.shape[0]}.")
print(f"The total number of columns in the filtered dataset is {hdi_2022_df.shape[1]}.")

# Displaying all columns and their datatypes
print("All columns and their datatypes : ")
print(hdi_2022_df.dtypes)
```

```python
# Task-3 : Missing Values & Data Cleaning

# Printing missing values in each column of the dataset
print("Total number of null values in each column : ")
print(hdi_2022_df.isnull().sum())

# There are no numeric columns stored as text and no inconsistent or misspelled country names

# Printing the number of duplicate rows
print(f"There are {hdi_2022_df.duplicated().sum()} duplicate rows.")

# Printing special characters representing missing data
print(f"There are {sum((hdi_2022_df.isin(['-','NA','N/A'])).sum())} values with special characters representing missing data.")

# Dropping rows with missing values
row_count_og=len(hdi_2022_df)
hdi_2022_df=hdi_2022_df.dropna()
row_count_new=len(hdi_2022_df)
print(row_count_og-row_count_new," rows with empty values dropped.\nEmpty rows are dropped as it protects data integrity by avoiding inaccurate data entry.")
```

```python
# Task-4 : Basic Statistics

# Displaying mean,median and standard deviation of the dataset
print("Mean of HDI for the year 2022 : ",hdi_2022_df['hdi'].mean())
print("Median of HDI for the year 2022 : ",hdi_2022_df['hdi'].median())
print("Standard Deviation of HDI for the year 2022 : ",hdi_2022_df['hdi'].std())

# Displaying country with highest HDI
max_country,max_hdi=hdi_2022_df.loc[hdi_2022_df["hdi"].idxmax(),["country","hdi"]]
print(f"Country with the highest HDI in 2022 is {max_country} with the HDI of {max_hdi}.")

# Displaying country with lowest HDI
min_country,min_hdi=hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin(),['country','hdi']]
print(f"Country with the lowest HDI in 2022 is {min_country} with the HDI of {min_hdi}.")
```

```python
# Task-5 : Filtering and Sorting

# Filtering the dataset to include countries with HDI over 0.8 only.
filtered_hdi_2022_df=hdi_2022_df[hdi_2022_df["hdi"] > 0.800]

# Sorting the dataset by GNI in descending order
hdi_2022_df_sorted_by_GNI=filtered_hdi_2022_df.sort_values(by="gross_inc_percap",ascending=False)

# Displaying the top 10 countries
print("Top 10 countries with highest Gross National Income (GNI) per Capita in the year 2022 : ")
display(hdi_2022_df_sorted_by_GNI[["country","gross_inc_percap"]].head(10))
```

```python
# Task-6 : Adding HDI Category Column

# Creating a new column and categorizing the entries into low,medium,high and very high
hdi_2022_df["HDI_Category"]=''
for index,row in hdi_2022_df.iterrows():
  hdi_value=row["hdi"]
  if hdi_value<=0.550 :
    hdi_2022_df.at[index,"HDI_Category"]="Low"
  elif hdi_value>0.550 and hdi_value<=0.699:
    hdi_2022_df.at[index,'HDI_Category']="Medium"
  elif hdi_value>0.699 and hdi_value<=0.799:
    hdi_2022_df.at[index,'HDI_Category']="High"
  else:
    hdi_2022_df.at[index,"HDI_Category"]="Very High"

# Displaying first five rows of the new dataset
print("First 5 rows of HDI dataset with added column : ")
display(hdi_2022_df.head())

# Saving the dataframe to a new dataset
hdi_2022_df.to_csv("/content/drive/MyDrive/5CS037--Concepts-and-Technologies-of-AI/CourseWork/HDI_category_added.csv",index=False)
print("New dataset created.")
```

```python
# 3.2 Problem- 1B- HDI Visualization and Trend Analysis (2020â€“ 2022)
# Task 1 : Data Extraction and Saving

# Filtering the dataframe to include only the years 2020,2021 and 2022
HDI_problem_1B=hdi_df[hdi_df["year"].isin([2020,2021,2022])]

# Saving the dataframe into a dataset
HDI_problem_1B.to_csv("/content/drive/MyDrive/5CS037--Concepts-and-Technologies-of-AI/CourseWork/HDI_problem_1B.csv",index=False)
print("Dataset created successfully.")
```

```python
# Task 2 : Data Cleaning

# Checking for missing values in hdi,country and year columns
print(f'There are {HDI_problem_1B["hdi"].isna().sum()} missing values in hdi column.')
print(f'There are {HDI_problem_1B["country"].isna().sum()} missing values in country column.')
print(f'There are {HDI_problem_1B["year"].isna().sum()} missing values in year column.\n')

# Checking for missing and duplicate values in the dataset
print(f"There are {HDI_problem_1B.isnull().sum().sum()} missing or null values.")
print(f"There are {HDI_problem_1B.duplicated().sum()} duplicate rows.")
print(f"There are {sum((HDI_problem_1B.isin(['-','NA','N/A'])).sum())} values with special characters such as '-' representing missing data.")

# Dropping rows with empty values
row_count_og=len(HDI_problem_1B)
HDI_problem_1B=HDI_problem_1B.dropna()
row_count_new=len(HDI_problem_1B)
print(row_count_og-row_count_new," rows with empty values dropped.")
```

```python
# Task_3 : Visualization Tasks
# A.Line Chart â€” HDI Trend (Country-Level)

# Selecting five countries and creating a new dataset with entries from that country only
df_for_visualilization=HDI_problem_1B[HDI_problem_1B["country"].isin(["Nigeria","United Kingdom","Nepal","Spain","Japan"])]

# Displaying year and hdi columns of the dataframe
display(df_for_visualilization[["country","year","hdi"]])

# Ploting HDI values for each country across the years 2020, 2021, and 2022.
plt.figure(figsize=(9,5))
sns.lineplot(data=df_for_visualilization,x="year",y="hdi",hue="country",marker='o')
plt.title("HDI Values across 2020,2021,2022 : ")
plt.xlabel("Year")
plt.ylabel("HDI")
plt.xticks([2020,2021,2022])
plt.legend(title="Country",bbox_to_anchor=(1.05,1),loc="upper left")
plt.figtext(0.5,-00.5,"Figure 1 :Line chart showing HDI values of five countries across a 3 year period.",wrap=True)
plt.tight_layout()
plt.show()
```

```python
# B.Bar Chart: Average HDI by Region (2020â€“2022)

# Grouping dataset by Country(Region) and Year and computing mean of HDI for each region-year pair
df_country_year=(df_for_visualilization.groupby(["country","year"],as_index=False)["hdi"].mean())

#Plotting bar chart comparing average HDI across regions for each year.
plt.figure(figsize=(10,6))
sns.barplot(data=df_country_year,x="country",y="hdi",hue="year",palette="colorblind")
plt.title("Average HDI by Region (2020â€“2022)")
plt.xlabel("Country")
plt.ylabel("HDI Value")
plt.ylim(0,1)
plt.legend(title="Year",bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```

```python
# B.Box Plot: HDI Distribution for 2020, 2021, and 2022

# Loading dataframe from problem 1B(Only years 2020,2021 and 2022)
df_2020_2022=HDI_problem_1B

# Creating a box plot showing HDI spread for each of the three years.
plt.figure(figsize=(10,6))
sns.boxplot(data=df_2020_2022,x="year",y="hdi",color="steelblue")
plt.title("HDI Distribution for 2020, 2021, and 2022")
plt.xlabel("Year")
plt.ylabel("HDI")
plt.ylim(0,1)
plt.show()
```

```python
# B. Scatter Plot : HDI vs. GNI per Capita

# Checking if the dataset contains a column for GNI per Capita.
if "gross_inc_percap" in HDI_problem_1B.columns:
  # Creating a scatter plot using HDI as the dependent variable and GNI per Capita on the x-axis.
    plt.figure(figsize=(10,6))
    plt.scatter(HDI_problem_1B["gross_inc_percap"],HDI_problem_1B["hdi"],alpha=0.6)
    plt.title("HDI vs. GNI per Capita")
    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.ylim(0,1)
    plt.show()
else:
    print("GNI per Capita variable not available in the dataset.")
```

```python
# Task-4 : Short Analysis Questions

# Which countries show the greatest improvement in HDI from 2020 to 2022?
hdi_2020_2022=HDI_problem_1B[HDI_problem_1B["year"].isin([2020,2021,2022])]
hdi_improvement=hdi_2020_2022.pivot(index="country",columns="year",values="hdi")
hdi_improvement["change_in_hdi_from_2020_to_2022"]=hdi_improvement[2022]-hdi_improvement[2020]
sorted_hdi_improvement=hdi_improvement.sort_values("change_in_hdi_from_2020_to_2022",ascending=False)
print("The countries which show the greatest improvement in HDI from 2020 to 2022 are : ")
display(sorted_hdi_improvement.head())

# Did any countries experience a decline in HDI? Provide possible reasons.
countries_with_HDI_decline=hdi_improvement[hdi_improvement["change_in_hdi_from_2020_to_2022"]<0]
countries_with_HDI_decline_df=countries_with_HDI_decline.reset_index()
print("The countries which displayed a decline in HDI are : ")
display(countries_with_HDI_decline_df[["country", "change_in_hdi_from_2020_to_2022"]])

# Which region has the highest and lowest average HDI across these three years?
average_hdi=hdi_improvement.mean(axis=1)
highest_avg=average_hdi.sort_values(ascending=False).head(1)
lowest_avg=average_hdi.sort_values().head(1)
print("The country with the highest average HDI across the three years:")
display(highest_avg)
print("The country with the lowest average HDI across the three years:")
display(lowest_avg)
```

```python
# 4. Problem 2 : Advanced HDI Exploration
# Task-1 : Create South Asia Subset

# Defining the list of south asian countries
south_asian_countries_list=["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]

# Filtering the HDI dataset to include only these countries.
south_asian_countries_df=HDI_problem_1B[HDI_problem_1B['country'].isin(south_asian_countries_list)]

# Saving the filtered dataset
south_asian_countries_df.to_csv("/content/drive/MyDrive/5CS037--Concepts-and-Technologies-of-AI/CourseWork/HDI_SouthAsia.csv",index=False)
print("South Asian Countries dataset created.")
```

```python
# Task-2 : Composite Development Score

# Creating composite score column

south_asian_countries_df=south_asian_countries_df.copy()
south_asian_countries_df["Composite_Score"]=(0.30*south_asian_countries_df['life_expectancy'] + 0.30*south_asian_countries_df['gross_inc_percap'])

# Rank South Asian countries based on Composite Score.
sorted_asian_countries_df=south_asian_countries_df.sort_values(by="Composite_Score",ascending=False)
sorted_asian_countries_df=sorted_asian_countries_df.drop_duplicates(subset="country", keep="first")
print("Top 5 countries based on their composite scores (Repeating countries not included) : ")
display(sorted_asian_countries_df[["country","year","Composite_Score"]].head())

# Plot the top 5 countries in a horizontal bar chart.
plt.figure(figsize=(10,6))
plt.barh(sorted_asian_countries_df["country"].head()+"("+sorted_asian_countries_df["year"].head().astype(str)+")",sorted_asian_countries_df["Composite_Score"].head())
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.gca().invert_yaxis()
plt.title("Top 5 South Asian Countries based on Composite Score")
plt.show()
```

```python
# Task-3 : Outlier Detection

# Function to detect outliers
def detect_outliers(column):
  Q1=column.quantile(0.25)
  Q3=column.quantile(0.75)
  IQR=Q3-Q1
  min_acceptable_val=Q1-1.5*IQR
  max_acceptable_val=Q3+1.5*IQR
  return (column<min_acceptable_val) | (column>max_acceptable_val)

# Detecting outliers in HDI and GNI per Capita
south_asian_countries_df["HDI_outlier"]=detect_outliers(south_asian_countries_df["hdi"])
south_asian_countries_df["GNI_outlier"]=detect_outliers(south_asian_countries_df["gross_inc_percap"])

south_asian_countries_df["Outlier"]="No Outlier"
south_asian_countries_df.loc[south_asian_countries_df["HDI_outlier"],"Outlier"]="HDI Outlier"
south_asian_countries_df.loc[south_asian_countries_df["GNI_outlier"], "Outlier"]="GNI Outlier"
south_asian_countries_df.loc[south_asian_countries_df["HDI_outlier"] & south_asian_countries_df["GNI_outlier"],"Outlier"]="HDI & GNI Outlier"

# Creating scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color
plt.figure(figsize=(10,6))
sns.scatterplot(data=south_asian_countries_df,x="gross_inc_percap",y="hdi",hue="Outlier",palette='colorblind')
plt.title("GNI per Capita vs HDI with Outliers Highlighted ")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.ylim(0,1)
plt.show()
```

```python
# Task-4 : Exploring Metric Relationships

# Selecting two HDI components (e.g., Gender Development Index and Life Expectancy Index {"life expectancy"})
hdi_components=["gender_development","life_expectancy"]

# Computing Pearson correlation of each metric with HDI.
for component in hdi_components:
    correlation=south_asian_countries_df[component].corr(south_asian_countries_df["hdi"])
    print(f"Pearson correlation between {component} and HDI: {correlation:.3f}")

# Creating scatter plots with trendlines to visualize the relationships.
    plt.figure(figsize=(10,6))
    sns.regplot(data=south_asian_countries_df,x=component,y="hdi",scatter_kws={"alpha":0.6},line_kws={"color":"red"})
    plt.title(f"HDI vs {component}")
    plt.xlabel(component)
    plt.ylabel("HDI")
    plt.ylim(0,1)
    plt.show()
```

```python
# Task-5 : Gap Analysis

south_asian_countries_df=south_asian_countries_df.copy()

# Creating a new column GNI_HDI_Gap
south_asian_countries_df["GNI_HDI_Gap"]=south_asian_countries_df["gross_inc_percap"]-south_asian_countries_df["hdi"]
sorted_south_asian_countries_df= south_asian_countries_df.groupby("country", as_index=False).agg({"GNI_HDI_Gap": "mean"})

# Ranking top and bottom 3 countries
top_positive_gaps=sorted_south_asian_countries_df.sort_values(by="GNI_HDI_Gap",ascending=False).head(3)
top_negative_gaps=sorted_south_asian_countries_df.sort_values(by="GNI_HDI_Gap",ascending=True).head(3)

# Plotting the top 3 positive gaps
plt.figure(figsize=(10,6))
plt.bar(top_positive_gaps["country"],top_positive_gaps["GNI_HDI_Gap"])
plt.xlabel("Country")
plt.ylabel("GNIâ€“HDI Gap")
plt.title("Top 3 Positive GNIâ€“HDI Gaps in South Asia")
plt.show()

# Plotting the top 3 negative gaps
plt.figure(figsize=(10,6))
plt.bar(top_negative_gaps["country"], top_negative_gaps["GNI_HDI_Gap"], color="red")
plt.xlabel("Country")
plt.ylabel("GNI â€“ HDI Gap")
plt.title("Top 3 Negative GNIâ€“HDI Gaps in South Asia")
plt.show()
```

```python
# 5 Problem-3 : Comparative Regional Analysis: South Asia vs Middle East
# Task-1 : Create Middle East Subset

# Defining the list of Middle East and South Asian Countries
south_asia_countries_list=["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]
middle_east_countries_list=["Bahrain","Iran","Iraq","Israel","Jordan","Kuwait","Lebanon","Oman","Palestine","Qatar","Saudi Arabia","Syria","United Arab Emirates","Yemen"]

# Creating subsets for South Asia and Middle Ease
south_asia_df=HDI_problem_1B[HDI_problem_1B["country"].isin(south_asia_countries_list)].copy()
middle_east_df=HDI_problem_1B[HDI_problem_1B["country"].isin(middle_east_countries_list)].copy()

# Saving the datasets
south_asia_df.to_csv("/content/drive/MyDrive/5CS037--Concepts-and-Technologies-of-AI/CourseWork/HDI_SouthAsia_2020_2022.csv",index=False)
middle_east_df.to_csv("/content/drive/MyDrive/5CS037--Concepts-and-Technologies-of-AI/CourseWork/HDI_MiddleEast_2020_2022.csv",index=False)
print("South Asia and Middle East subsets saved successfully.")
```

```python
# Task-2 : Descriptive Statistics

# Computing mean and standard deviation of HDI for each region (South Asia vs Middle East)
mean_south_asia=south_asia_df["hdi"].mean()
mean_middle_east=middle_east_df["hdi"].mean()
sd_south_asia=south_asia_df["hdi"].std()
sd_middle_east=middle_east_df["hdi"].std()

print("HDI mean of South Asia dataset",mean_south_asia)
print("HDI mean of Middle East dataset",mean_middle_east)
print("HDI standard deviation of South Asia dataset",sd_south_asia)
print("HDI standard deviation of Middle East  dataset",sd_middle_east)

# Identifying which region performs better on average.
if(mean_south_asia>mean_middle_east):
  print("South Asia performs better on average.")
else:
  print("Middle East performs better on average.")
```

```python
# Task-3 : Top and Bottom Performers

#Identifying top 3 and bottom 3 countries in each region based on HDI
grouped_south_asia_df=south_asia_df.groupby("country",as_index=False)["hdi"].mean()
grouped_middle_east_df=middle_east_df.groupby("country",as_index=False)["hdi"].mean()

print("Top 3 countries in South Asia : ")
display(grouped_south_asia_df.sort_values(by="hdi",ascending=False).head(3))
print("Top 3 countries in Middle East : ")
display(grouped_middle_east_df.sort_values(by="hdi",ascending=False).head(3))
print("Bottom 3 countries in South Asia : ")
display(grouped_south_asia_df.sort_values(by="hdi",ascending=True).head(3))
print("Bottom 3 countries in Middle East : ")
display(grouped_middle_east_df.sort_values(by="hdi",ascending=True).head(3))


# Creating  a bar chart comparing these top and bottom performers across the two regions.
plt.figure(figsize=(10,6))
combined_df_for_plot=pd.concat([grouped_south_asia_df.sort_values(by="hdi",ascending=False).head(3).assign(Region="South Asia",Type="Top"),
                                grouped_middle_east_df.sort_values(by="hdi",ascending=False).head(3).assign(Region="Middle East",Type="Bottom"),
                                grouped_south_asia_df.sort_values(by="hdi",ascending=True).head(3).assign(Region="South Asia",Type="Top"),
                                grouped_middle_east_df.sort_values(by="hdi",ascending=True).head(3).assign(Region="Middle East",Type="Bottom")])
sns.barplot(data=combined_df_for_plot,x="country",y="hdi",hue="Region",palette="colorblind")
plt.title("Top and Bottom Performers by HDI")
plt.ylabel("Average HDI")
plt.xticks(rotation=45)
plt.show()
```

```python
# Task-4 : Metric Comparisons

# Defining the metrics
metrics=["gender_development","life_expectancy","gross_inc_percap"]

# Calculating mean values for each region
south_asia_means=south_asia_df.mean(numeric_only=True)
middle_east_means=middle_east_df.mean(numeric_only=True)

# Plotting group bar charts
fig,axes=plt.subplots(1,3,figsize=(10,6))
for ax,metric in zip(axes,metrics):
    data=pd.DataFrame({"Region":["South Asia","Middle East"],"Value":[south_asia_means[metric],middle_east_means[metric]]})
    sns.barplot(data=data,x="Region", y="Value",ax=ax)
    ax.set_title(metric)
    ax.set_ylabel("Average Value")
    ax.set_xlabel("Region")

plt.suptitle("Comparison of HDI Component Metrics Across Regions")
plt.show()
```

```python
# Task-5 : HDI Disparity

# Computing the range (maxâ€“ min) of HDI for each region.
range_south_asia=south_asia_df["hdi"].max()-south_asia_df["hdi"].min()
range_middle_east=middle_east_df["hdi"].max()-middle_east_df["hdi"].min()
print("Range of HDI for the South Asia Region : ",range_south_asia)
print("Range of HDI for the Middle East Region : ",range_middle_east)

# Computing the coefficient of variation (CV = std/mean) for HDI.
cv_east_asia=south_asia_df["hdi"].std()/south_asia_df["hdi"].mean()
cv_middle_east=middle_east_df["hdi"].std()/middle_east_df["hdi"].mean()
print("Coefficient of Variation of HDI for the South Asia Region : ",cv_east_asia)
print("Coefficient of Variation of HDI for the Middle East Region : ",cv_middle_east)
```

```python
# Task-6 : Correlation Analysis

# Metrics for correlation analysis
metrics=["gender_development","life_expectancy"]

for region_name,df in [("South Asia",south_asia_df),("Middle East",middle_east_df)]:
    print(f"\n{region_name} correlations:")
    for metric in metrics:
      #Computing correlations
        correlation=df[metric].corr(df["hdi"])
        print(f"{metric} vs HDI: {correlation:.3f}")

        # Creating scatterplots with trendlines
        plt.figure(figsize=(10,6))
        sns.regplot(data=df,x=metric,y="hdi",scatter_kws={"alpha":0.6},line_kws={"color":"red"})
        plt.title(f"{metric} vs HDI of ({region_name})")
        plt.xlabel(metric)
        plt.ylabel("HDI")
        plt.ylim(0,1)
        plt.show()
```

```python
# Task-7 : Outlier Detection

# Function to detect outliers
def detect_outliers(column):
  Q1=column.quantile(0.25)
  Q3=column.quantile(0.75)
  IQR=Q3-Q1
  min_acceptable_val=Q1-1.5*IQR
  max_acceptable_val=Q3+1.5*IQR
  return (column<min_acceptable_val) | (column>max_acceptable_val)

# Detecting outliers in HDI and GNI per Capita
south_asia_df["HDI_outlier"]=detect_outliers(south_asia_df["hdi"])
south_asia_df["GNI_outlier"]=detect_outliers(south_asia_df["gross_inc_percap"])
middle_east_df["HDI_outlier"]=detect_outliers(middle_east_df["hdi"])
middle_east_df["GNI_outlier"]=detect_outliers(middle_east_df["gross_inc_percap"])

south_asia_df["Outlier"]="No Outlier"
south_asia_df.loc[south_asia_df["HDI_outlier"],"Outlier"]="HDI Outlier"
south_asia_df.loc[south_asia_df["GNI_outlier"], "Outlier"]="GNI Outlier"
south_asia_df.loc[south_asia_df["HDI_outlier"] & south_asia_df["GNI_outlier"],"Outlier"]="HDI & GNI Outlier"

middle_east_df["Outlier"]="No Outlier"
middle_east_df.loc[middle_east_df["HDI_outlier"],"Outlier"]="HDI Outlier"
middle_east_df.loc[middle_east_df["GNI_outlier"], "Outlier"]="GNI Outlier"
middle_east_df.loc[middle_east_df["HDI_outlier"] & middle_east_df["GNI_outlier"],"Outlier"]="HDI & GNI Outlier"

# Creating scatter plot of GNI per Capita vs HDI for each region, highlighting the outliers in a different color
plt.figure(figsize=(10,6))
sns.scatterplot(data=south_asia_df,x="gross_inc_percap",y="hdi",hue="Outlier",palette='colorblind')
plt.title("GNI per Capita vs HDI with Outliers Highlighted for South Asia Region ")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.ylim(0,1)
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(data=middle_east_df,x="gross_inc_percap",y="hdi",hue="Outlier",palette='colorblind')
plt.title("GNI per Capita vs HDI with Outliers Highlighted for Middle East Region ")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.ylim(0,1)
plt.show()
```
